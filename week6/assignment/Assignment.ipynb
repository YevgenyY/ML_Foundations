# Split SFrame into 4 categories

```python
dog = image_train.filter_by('dog', 'label')
cat = image_train.filter_by('cat', 'label')
car = image_train.filter_by('automobile', 'label')
bird = image_train.filter_by('bird', 'label')
```

# Create knn models for each category

```python
dog_model = graphlab.nearest_neighbors.create(dog,features=['deep_features'],
                                             label='id')
cat_model = graphlab.nearest_neighbors.create(cat,features=['deep_features'],
                                             label='id')
car_model = graphlab.nearest_neighbors.create(car,features=['deep_features'],
                                             label='id')
bird_model = graphlab.nearest_neighbors.create(bird,features=['deep_features'],
                                             label='id')
```

# Create lambdas for images retrieval

```python
graphlab.canvas.set_target('ipynb')
def get_images_from_ids(query_result):
    return image_train.filter_by(query_result['reference_label'],'id')

show_dogs = lambda i: get_images_from_ids(dog_model.query(image_train[i:i+1]))['image'].show()
show_cats = lambda i: get_images_from_ids(cat_model.query(image_train[i:i+1]))['image'].show()
show_cars = lambda i: get_images_from_ids(car_model.query(image_train[i:i+1]))['image'].show()
show_birds = lambda i: get_images_from_ids(bird_model.query(image_train[i:i+1]))['image'].show()
```

#  Computing nearest neighbors accuracy using SFrame operations: A nearest 

```python
image_test_dog = image_test.filter_by('dog', 'label')
image_test_cat = image_test.filter_by('cat', 'label')
image_test_car = image_test.filter_by('automobile', 'label')
image_test_bird = image_test.filter_by('bird', 'label')
```

# Find distances between Dogs and Cats
The question is :<BR>
How many cat images (testing set) are closer to dog images (training set) than dog images from testing set to dog images (training set).

```python
dog_dog_neighbors = dog_model.query(image_test_dog, k=1)
dog_cat_neighbors = cat_model.query(image_test_dog, k=1)
dog_car_neighbors = car_model.query(image_test_dog, k=1)
dog_bird_neighbors = bird_model.query(image_test_dog, k=1)
dog_dog_neighbors
```

```python
dog_cat_neighbors
```

## Calculate distances between dog vs others images

```python
dog_distances = graphlab.SFrame({'dog-dog':dog_dog_neighbors['distance'], 
                                 'dog-cat':dog_cat_neighbors['distance'],
                                 'dog-car':dog_car_neighbors['distance'],
                                 'dog-bird':dog_bird_neighbors['distance'],
                                 })
```

```python
dog_distances.head()
```

## Check predictions correctness for each dog image

```python
def is_dog_correct(row):
    rc = 1;
    if row['dog-dog'] > row['dog-cat'] or row['dog-dog'] > row['dog-car'] or row['dog-bird'] > row['dog-car']:
        rc = 0;
    
    return rc;
```

```python
is_dog_correct(dog_distances[0])
#dog_distances[0]
```

```python
dog_distances['is_correct'] = dog_distances.apply(is_dog_correct)
```

```python
dog_distances['is_correct'].sum()
```

```python
dog_distances['is_correct'].size()
```

## Calculate the accuracy

```python
ALL = dog_distances['is_correct'].size()
TP = dog_distances['is_correct'].sum()
FP = TN = 0
FN = ALL - TP

# Accuracy
float(TP + TN) / ALL 
```
